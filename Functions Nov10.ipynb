```python
# Exercise 05

def my_abs(number):
    """Return the absolute value of a number if it's int or float, else raise a TypeError."""
    # Type check using isinstance()
    if not isinstance(number, (int, float)):
        print("Error: Input must be an integer or float!")
        return None

    # Calculate absolute value manually
    if number < 0:
        number = -number
    return number


# --- Testing the function ---
print("my_abs(-10) =", my_abs(-10))     
print("my_abs(7.5) =", my_abs(7.5))      
print("my_abs(0) =", my_abs(0))        
print("my_abs('Babson') =", my_abs('Babson'))
```

```python
# Exercise 06

import math

def quadratic(a, b, c):
    """Solve a quadratic equation ax² + bx + c = 0 and return its two roots."""
    # Calculate discriminant
    d = b ** 2 - 4 * a * c
    
    # Check if discriminant is negative
    if d < 0:
        return None, None  # No real roots
    
    # Calculate both roots
    root1 = (-b + math.sqrt(d)) / (2 * a)
    root2 = (-b - math.sqrt(d)) / (2 * a)
    
    return root1, root2


# --- Testing the function ---
x1, x2 = quadratic(1, -3, 2)   # x² - 3x + 2 = 0 → roots: 2 and 1
print("Roots for (1, -3, 2):", x1, x2)

x1, x2 = quadratic(1, 2, 1)    # x² + 2x + 1 = 0 → one root: -1
print("Roots for (1, 2, 1):", x1, x2)

x1, x2 = quadratic(1, 0, 1)    # x² + 1 = 0 → no real roots
print("Roots for (1, 0, 1):", x1, x2)
```
