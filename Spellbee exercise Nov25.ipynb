```python
#spellbee exercise - Dear professor, please note I did this with the help of AI. 

# 0. Sample dictionary (different words from the earlier example)
word_list = [
    "stone", "tones", "notes", "nest", "sent",
    "tension", "tiny", "seton", "sonnet", "section",
    "note", "tone", "one", "site", "siteon"
]

# 1. Get puzzle letters from the user 
core = input("Enter the center letter: ").lower()
ring = input("Enter the other six letters: ").lower()

# 2. Build the set/string of allowed letters 
letters_allowed = core + ring

# 3. Check each candidate word against the rules
solutions = []

for candidate in word_list:
    w = candidate.lower()

    # Rule 1: word must be at least 4 letters long
    if len(w) < 4:
        continue

    # Rule 2: word must contain the center letter at least once
    if core not in w:
        continue

    # Rule 3: every letter must come from the allowed set
    ok = True
    for ch in w:
        if ch not in letters_allowed:
            ok = False
            break

    if ok:
        solutions.append(w)

# 4. Show the results
print("\nValid Spelling Bee words:")
for w in solutions:
    print("-", w)

print(f"\nTotal words found: {len(solutions)}")
```
