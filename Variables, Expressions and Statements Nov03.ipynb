```python
#exercise 2 - question 2

cover_price = 24.95
discount = 0.4
copies = 60

# Discounted price as requested
discounted_price = cover_price * (1 - discount)

# Total cost of books calculation
book_cost = discounted_price * copies

# Overall Shipping cost
shipping_cost = 3 + 0.75 * (copies - 1)

# Total wholesale cost
total_cost = book_cost + shipping_cost

print("Total wholesale cost for 60 copies is: $", total_cost)
```

```python
#exercise 2 - question 3

total = 750
drank = 250

remaining = total - drank
percent_remaining = (remaining / total) * 100

print(f"{percent_remaining:.1f}%")
```
