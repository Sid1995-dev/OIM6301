```python
#exercise 01 - Fermat's Last Theorem part

def check_fermat(a, b, c, n):
    """Checks whether a, b, c, n violate Fermat's Last Theorem."""
    if n > 2 and (a ** n + b ** n == c ** n):
        print("Holy smokes, Fermat was wrong!")
    else:
        print("No, that doesn't work.")

def prompt_fermat():
    """Prompts the user to input values for a, b, c, and n, converts them to integers,
    and checks Fermatâ€™s theorem."""
    a = int(input("Enter value for a: "))
    b = int(input("Enter value for b: "))
    c = int(input("Enter value for c: "))
    n = int(input("Enter value for n: "))
    check_fermat(a, b, c, n)

prompt_fermat()
```

```python
#exercise 01 - BMI Calculator part

#here's the BMI Calculator - 

def calculate_bmi(weight, height):
    """Returns the BMI given weight (kg) and height (m)."""
    bmi = weight / (height ** 2)
    return bmi

def get_bmi_category():
    """Prompts user for weight and height, calculates BMI, and prints category."""
    weight = float(input("Enter your weight in kg: "))
    height = float(input("Enter your height in meters: "))

    bmi = calculate_bmi(weight, height)
    print(f"Your BMI is: {bmi:.2f}")

    if bmi < 18.5:
        print("Category: Underweight")
    elif bmi < 25:
        print("Category: Normal weight")
    elif bmi < 30:
        print("Category: Overweight")
    else:
        print("Category: Obesity")

get_bmi_category()
```

```python
#exercise 02 - factorial exercise

def factorial(n: int) -> int:
    """Return n! recursively. Requires n >= 0."""
    if n < 0:
        raise ValueError("factorial is undefined for negative n")
    if n <= 1:
        return 1
    return n * factorial(n - 1)

# quick checks, checking the factorial exercise here -
print("factorial(0) =", factorial(0))  
print("factorial(1) =", factorial(1)) 
print("factorial(5) =", factorial(5))
```

```python
#exercise 02 - fibonacci exercise

def fibonacci(n: int) -> int:
    """
    Return the n-th Fibonacci number recursively with F(0)=0, F(1)=1.
    Requires n >= 0.
    """
    if n < 0:
        raise ValueError("fibonacci requires n >= 0")
    if n < 2:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)

# quick checks, testing the fibonacci exercise - 
print("fibonacci(0) =", fibonacci(0)) 
print("fibonacci(1) =", fibonacci(1)) 
print("fibonacci(10) =", fibonacci(10))
```

```python
#exercise 02 - greatest common divisor

def gcd(a, b):
    """Return the greatest common divisor (GCD) of two positive integers using an iterative approach."""
    if a <= 0 or b <= 0:
        raise ValueError("Both numbers must be positive integers")

    # Iterative Euclidean algorithm
    while b != 0:
        a, b = b, a % b
    return a

# --- Testing the functions here ---
print("gcd(2, 12) =", gcd(2, 12))     
print("gcd(6, 12) =", gcd(6, 12))     
print("gcd(9, 12) =", gcd(9, 12))    
print("gcd(17, 12) =", gcd(17, 12))
```
