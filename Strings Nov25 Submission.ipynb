```python
# Strings - using len function

# Counting the number of characters in the string
team = "New England Patriots"

print(len(team))
```

```python
# Strings - Traversal with a for loop function - traversing using a while loop

# Going through the string one character at a time using a while loop
team = "New England Patriots"

idx = 0
while idx < len(team):
    print(team[idx])
    idx += 1
```

```python
# Looping directly through each character in the string - traversing using a for loop
for ch in team:
    print(ch)
```

```python
# Strings - Exercise 01 

# here we are Fixing Ouack and Quack

prefixes = "JKLMNOPQ"
ending = "ack"

for letter in prefixes:
    if letter == "O":
        print("Ou" + ending)     # Ouack
    elif letter == "Q":
        print("Qu" + ending)     # Quack
    else:
        print(letter + ending)
```

```python
# Exercise 02 — Generalized counting function and string methods

def count(text, target_letter):
    """Return the number of times target_letter appears in text."""
    total = 0
    for char in text:
        if char == target_letter:
            total += 1
    return total

# Test the function
print(count('New England Patriots', 'a'))
print(count('banana', 'n'))
print(count('Hello World', 'o'))
```

```python
# Exercise 03 - The in_both function

def in_both(word1, word2):
    for letter in word1:
        if letter in word2:
            print(letter)

# Test run
in_both("New", "England")
```

```python
# Exercise 03 – experimenting with string methods in this section

text = "   Hello there! Welcome to Python.   "

# strip(): removes spaces at start & end
cleaned_text = text.strip()
print("strip():", cleaned_text)

# split(): breaks into a list of words
words_list = cleaned_text.split()
print("split():", words_list)

# replace(): substitutes text
replaced_text = cleaned_text.replace("Python", "Programming")
print("replace():", replaced_text)

# optional: combining them
combo = text.strip().replace("Hello", "Hi").upper()
print("combined:", combo)
```

```python
# Exercise 04 - 
def any_lowercase_1(s):
    for c in s:
        if c.islower():
            return True
        else:
            return False

def any_lowercase_2(s):
    for c in s:
        if 'c'.islower():
            return 'True'
        else:
            return 'False'

def any_lowercase_3(s):
    for c in s:
        flag = c.islower()
    return flag

def any_lowercase_4(s):
    flag = False
    for c in s:
        flag = flag or c.islower()
    return flag

def any_lowercase_5(s):
    for c in s:
        if not c.islower():
            return False
    return True
```

```python
# Exercise 04 – Explanation for above.

**any_lowercase_1(s)**  
Checks only the *first character*. Returns True if the first character is lowercase, otherwise False.

**any_lowercase_2(s)**  
Does not check the string at all. It tests the literal character `'c'`, so it always returns the *string* `'True'`.

**any_lowercase_3(s)**  
Returns whether the *last character* in the string is lowercase. Ignores the earlier characters.

**any_lowercase_4(s)**  
Correct function. Returns True if *any* character in the string is lowercase.

**any_lowercase_5(s)**  
Returns True only if *every* character is lowercase. So it checks “all lowercase,” not “any lowercase.”
```

```python
# Exercise 05 – Rotate word (Caesar cipher)

def rotate_word(word, n):
    rotated = ""

    for char in word:
        # Process lowercase letters
        if 'a' <= char <= 'z':
            start = ord('a')
            offset = (ord(char) - start + n) % 26
            rotated += chr(start + offset)

        # Process uppercase letters
        elif 'A' <= char <= 'Z':
            start = ord('A')
            offset = (ord(char) - start + n) % 26
            rotated += chr(start + offset)

        # Leave non-letters unchanged
        else:
            rotated += char

    return rotated


# Example tests (you can keep or remove)
print(rotate_word("cheer", 7))     # - jolly
print(rotate_word("melon", -10))   # - cubed
print(rotate_word("HAL", -1))      # - IBM
```
