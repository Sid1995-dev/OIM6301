```python
#Iterations Exercise 03: Newton's Method for Square Roots

def mysqrt(a):
    x = a / 2        # initial estimate
    epsilon = 0.0000001
    
    while True:
        y = (x + a/x) / 2     # Newton's method update
        if abs(y - x) < epsilon:   # stop when close enough
            return y
        x = y
```

```python
import math

def test_square_root():
    print("a\tmysqrt(a)\tmath.sqrt(a)\tdiff")
    print("-\t---------\t------------\t----")
    
    for a in range(1, 10):
        my = mysqrt(a)
        actual = math.sqrt(a)
        diff = abs(my - actual)
        print(f"{a}\t{my}\t{actual}\t{diff}")
```
